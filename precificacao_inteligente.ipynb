```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('/content/hospedagem.json')
df.head()
```

NORMALIZAÇÃO

```python
df1 = pd.json_normalize(df['info_moveis'])
df1
```

```python
colunas = list(df1.columns)
colunas
```

```python
dados = df1.explode(colunas[3:]) #explode tira os dados das linhas de dentro das listas
dados.reset_index(inplace = True, drop = True)
dados
```

```python
dados.info()
```

```python
 dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

```python
dados.info()
```

```python
dados_inteiros = ['quantidade_banheiros',	'quantidade_quartos',	'quantidade_camas']
dados[dados_inteiros] = dados[dados_inteiros].astype(np.int64)
dados.info()
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
dados.info()
```

```python
#as colunas taxa_deposito,	taxa_limpeza e preco, tem caracteres estranhos, por isso precisam ser alterados
colunas_preco = ['taxa_deposito',	'taxa_limpeza',	'preco']
dados[colunas_preco] = dados[colunas_preco].applymap(lambda x: x.replace('$', '').replace(',', '').strip()) #strip apaga espaços em branco antes e depois, o applymap, serve
dados[colunas_preco]                                                                                        # que a função lambda percorra todas as colunas especificadas.
```

```python
dados[colunas_preco] = dados[colunas_preco].astype(np.float64)
dados.info()
```

tokenização

```python
dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True) #seleciona todos os dados que n contenham o que esta no regex
dados['descricao_local']
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', '', regex=True) #seleciona ifens que estão soltos
dados['descricao_local']
```

```python
dados['descricao_local'] = dados['descricao_local'].str.split()
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"','',regex=True)
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower()
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split()
```
