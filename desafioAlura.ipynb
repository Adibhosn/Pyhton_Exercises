```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
alunos = pd.read_csv(url)
alunos.head(5)
```

```python
alunos['pontos_extras'] = alunos['Notas'] * 1.4 - alunos['Notas']
alunos
```

```python
alunos['Notas_finais'] = alunos['Notas'] + alunos['pontos_extras']
alunos
```

```python
alunos['Aprovados'] = alunos['Notas_finais'].apply(lambda x: 'Aprovado' if x >= 6 else 'Reprovado')
alunos
```

```python
alunos.query('Notas < 6 & Notas_finais > 6')
```
