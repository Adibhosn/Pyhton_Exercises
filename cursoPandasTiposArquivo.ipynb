Importante: Na aula nós realizamos a normalização em um arquivo com o formato JSON. No entanto, a função json_normalize() aceita apenas um dicionário ou uma lista de dicionários. Por isso, no vídeo foi usada a estratégia de usar o código: pd.json_normalize(dados_pacientes_2['Pacientes']). Porém, se tentarmos usar parâmetros da função json_normalize em um arquivo JSON podem surgir erros. Para contornar isso, precisamos importar o módulo json e ler os arquivos conforme o código abaixo:

```python
#Importando o módulo JSON
import json

#importando requests
import requests

Pacientes = requests.get('https://raw.githubusercontent.com/alura-cursos/Pandas/main/pacientes_2.json')
#Lendo o arquivo json usando o módulo Python JSON
with open(Pacientes,'r') as f:
    dados = json.loads(f.read())

#Normalizando os dados com os parâmetros record_path e meta
pd.json_normalize(dados, record_path='Pacientes', meta=['Pesquisa', 'Ano'])


#o erro se deve ao fato de eu não ter carregado o arquivo para o meu dispositivo
```

```python
import json
import requests

# Fazendo a requisição para obter o conteúdo do arquivo JSON
response = requests.get('https://raw.githubusercontent.com/alura-cursos/Pandas/main/pacientes_2.json')

# Lendo o conteúdo JSON da resposta
dados = response.json()

# Normalizando os dados com os parâmetros record_path e meta
df = pd.json_normalize(dados, record_path='Pacientes', meta=['Pesquisa', 'Ano'])

# Exibindo o DataFrame resultante
df.head()

#USANDO O REQUESTS PARA CONSUMIR DADOS DE UMA API
# if response.status_code == 200: (ESSE TRECHO DE CODIGO VERIFICA SE A REQUISISIÇAO FOI BEM SUCESSIDA)
```
